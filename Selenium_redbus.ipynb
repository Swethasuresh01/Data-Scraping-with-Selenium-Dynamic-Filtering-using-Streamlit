```python
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException, ElementClickInterceptedException
import time

wait = WebDriverWait(driver, 20)

def close_overlays():
    try:
        overlay_close_button = driver.find_element(By.XPATH, "//button[@class='close']")
        overlay_close_button.click()
    except NoSuchElementException:
        pass  # No overlay found

def Kerala_link_route(path):
    LINKS_KERALA = []
    ROUTE_KERALA = []
    
    for i in range(1, 3):  # Adjust range for more pages
        try:
            paths = wait.until(EC.presence_of_all_elements_located((By.XPATH, path)))
            
            for links in paths:
                d = links.get_attribute("href")
                LINKS_KERALA.append(d)
                
            for route in paths:
                ROUTE_KERALA.append(route.text)
            
            # Attempt to locate and click the next button
            pagination = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@class="DC_117_paginationTable"]')))
            next_button = pagination.find_element(By.XPATH, f'//div[@class="DC_117_pageTabs " and text()={i+1}]')
            
            # Close any overlays before clicking
            close_overlays()

            # Use JavaScript click to avoid interception
            driver.execute_script("arguments[0].scrollIntoView(true);", next_button)
            time.sleep(1)  # Ensure scroll settles
            driver.execute_script("arguments[0].click();", next_button)
        
        except ElementClickInterceptedException:
            print("Element click intercepted again, retrying with JavaScript click after checking overlays.")
            close_overlays()  # Retry closing overlay if it appears again
            driver.execute_script("arguments[0].click();", next_button)
        
        except (NoSuchElementException, TimeoutException):
            print(f"No more pages to paginate at step {i}")
            break

    return LINKS_KERALA, ROUTE_KERALA

LINKS_KERALA, ROUTE_KERALA = Kerala_link_route("//a[@class='route']")
```

```python
df_k = pd.DataFrame({"Route_name" : ROUTE_KERALA, "Route_link": LINKS_KERALA})
df_k
```

```python
path = r"E:/Redbus_Project1/df_K.csv"
df_k.to_csv(path,index=False)
```

```python
from selenium.common.exceptions import NoSuchWindowException, TimeoutException

wait = WebDriverWait(driver, 20)

def check_window_open(driver):
    try:
        driver.title  # Attempt to access the title to check if the window is open
        return True
    except NoSuchWindowException:
        return False

def Andhra_link_route(path):
    if not check_window_open(driver):
        print("Browser window is closed. Restart the driver.")
        # Optionally, reinitialize the driver here if you want to continue

    LINKS_ANDHRA = []
    ROUTE_ANDHRA = []

    for i in range(1, 5):
        if not check_window_open(driver):
            print("Browser window was closed during execution.")
            break

        try:
            paths = wait.until(EC.presence_of_all_elements_located((By.XPATH, path)))

            for links in paths:
                d = links.get_attribute("href")
                LINKS_ANDHRA.append(d)

            for route in paths:
                ROUTE_ANDHRA.append(route.text)

            pagination = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@class="DC_117_paginationTable"]')))
            next_button = pagination.find_element(By.XPATH, f'//div[@class="DC_117_pageTabs " and text()={i+1}]')

            close_overlays()
            driver.execute_script("arguments[0].scrollIntoView(true);", next_button)
            time.sleep(1)
            driver.execute_script("arguments[0].click();", next_button)

        except (NoSuchElementException, TimeoutException):
            print(f"No more pages to paginate at step {i}")
            break

    return LINKS_ANDHRA, ROUTE_ANDHRA

LINKS_ANDHRA, ROUTE_ANDHRA = Andhra_link_route("//a[@class='route']")
```
